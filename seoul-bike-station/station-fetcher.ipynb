```python
import pandas as pd
bike = pd.read_csv('bike.csv', encoding='cp949')
bike_id = bike['대여소\n번호']
bike_x = bike['위도(Y)']
bike_y = bike['경도(X)']
bike
```

```python
from pyproj import Transformer
from shapely.geometry import Point as point
import geopandas as gpd
from tqdm import tqdm

points = []

for i in tqdm(range(len(bike)), desc="Fetching Stations"):
    idx = bike_id[i]
    points.append(point(bike_y[i], bike_x[i]))

points = gpd.GeoDataFrame(geometry=points, crs={'init': 'epsg:4326'})
points.to_file('shp/seoul-bike-stations.shp', driver="Shapefile")
```
