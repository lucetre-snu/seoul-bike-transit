```python
import demo
import warnings
from IPython.display import HTML
from google.colab import data_table
from bs4 import BeautifulSoup

def optimize_routes(org, des):
    with warnings.catch_warnings():
        warnings.simplefilter("ignore", category=RuntimeWarning)
        return demo.read_routes(org, des)

def render_dtbl(df_org):
    df = df_org.copy()
    for col in 'srt_tran end_tran srt_calo end_calo bike_calo'.split():
        try: del df[col]
        except: pass
    for id, row in df.iterrows():
        df.loc[id, 'org'] = BeautifulSoup(row['org']).text
        df.loc[id, 'des'] = BeautifulSoup(row['des']).text
        df.loc[id, 'srt_name'] = BeautifulSoup(row['srt_name']).text
        df.loc[id, 'end_name'] = BeautifulSoup(row['end_name']).text
    return data_table.DataTable(df, include_index=False, num_rows_per_page=10)

def render_html(df_org, by="time", ascending=True):
    df = df_org.copy()
    for col in df.columns:
        if col.find('tran') != -1 or col.find('time') != -1 or col.find('dist') != -1 or col.find('fare') != -1 or col.find('calo') != -1:
            if col.find(by) == -1 and col.find('total') == -1:
                try: del df[col]
                except: pass
    return HTML(df.sort_values(by=f'total_{by}', ascending=ascending).head().to_html(escape=False))

df = optimize_routes('샤로수길', '녹두거리')
# df = optimize_routes('대한건축학회', '서울대학교 자하연')
# df = optimize_routes('반포한강공원', '여의도한강공원')
# df = optimize_routes('낙성대공원', '몰로코')
# df = optimize_routes('삼성고', '서울신림초')
# df = optimize_routes('서울시청', '서울고속터미널')
# df = optimize_routes('건대입구', '쥬티클럽')
```

```python
render_dtbl(df)
```

```python
render_html(df, by='time', ascending=True)
```
